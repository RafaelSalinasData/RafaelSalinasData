```python
#CSV "fifa_eda.csv"

# Lista de diccionarios
df_fifa = pd.read_csv('fifa_eda.csv')
fifa_lista = df_fifa.to_dict(orient='records')

# las 10 primeras filas
primeras_filas_fifa = df_fifa.head(10)
print("Primeras 10 filas - FIFA:")
print(primeras_filas_fifa)

# las 10 últimas filas
ultimas_filas_fifa = df_fifa.tail(10)
print("\nÚltimas 10 filas - FIFA:")
print(ultimas_filas_fifa)

# Seleccionar 4 columnas y guardar en nuevo archivo
fifa_reducido = df_fifa[['Name', 'Overall', 'Club', 'Nationality']]
fifa_reducido.to_csv('fifa_reducido.csv', index=False)
print("\nArchivo 'fifa_reducido.csv' generado con éxito.")
```

```python
import pandas as pd
import json

#  JSON - "season1.json" 

# Leer archivo y guardarlo en una lista
with open('season1.json', 'r', encoding='utf-8') as file:
    got_data = json.load(file)

# Convertir a DataFrame
df_got = pd.DataFrame(got_data)

# las 10 primeras filas
primeras_filas_got = df_got.head(10)
print("\nPrimeras 10 filas - Subtítulos Season 1:")
print(primeras_filas_got)

# las 10 últimas filas
ultimas_filas_got = df_got.tail(10)
print("\nÚltimas 10 filas - Subtítulos Season 1:")
print(ultimas_filas_got)

# dame 2 columnas y guardar en un nuevo archivo JSON
if 'speaker' in df_got.columns and 'text' in df_got.columns:
    got_reducido = df_got[['speaker', 'text']]
    got_reducido.to_json('season1_reducido.json', orient='records', indent=4, force_ascii=False)
    print("\nArchivo 'season1_reducido.json' generado con éxito.")
else:
    print("\nNo se encontraron las columnas 'speaker' y 'text'. Verifica el contenido del archivo JSON.")
```
