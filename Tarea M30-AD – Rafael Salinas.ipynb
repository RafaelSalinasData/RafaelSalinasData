```python
df['Release Clause'] = pd.to_numeric(df['Release Clause'], errors='coerce')
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
df['Wage'] = pd.to_numeric(df['Wage'], errors='coerce')

# Revisar
df[['Value', 'Wage', 'Release Clause']].describe()
```

```python
current_year = datetime.now().year
df['Years Playing'] = current_year - df['Joined']

df[['Name', 'Joined', 'Years Playing']].head(10)
```

```python
df[df['Nationality'] == 'Mexico']
```

```python
df_sorted = df.sort_values(by='Release Clause', ascending=False)
df_sorted[['Name', 'Release Clause']].head(10)
```

```python
joined_group = df.groupby('Joined').size().reset_index(name='NumPlayers')
joined_group.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(joined_group['Joined'], joined_group['NumPlayers'], marker='o', color='teal')
plt.title('Número de jugadores por año de Joined')
plt.xlabel('Año Joined')
plt.ylabel('Número de jugadores')
plt.grid(True)
plt.show()
```
